### Q1. 편향은 어떤 기능을 수행하나요?  
- 세타는 실수값을 나타냄. 입력된 가중치의 합이 뉴런의 활성도를 나타내는 것.+b를 해줌으로써 좀더 뉴런이 활성화를 촉진시켜주는 존재? 
- 활성화 함수의 활성빈도 설정. 편향의 값이 커질수록 역치값이 커진다. +값을 줄수록 활성화가 더 됨. -값일때 활성화가 덜됨. 
- 활성화 함수의 활성 빈도를 설정할 때 사용합니다.
- 편향의 값이 높을 수록 활성화 함수의 역치 값이 커집니다.
- 편향의 기능 = 용도. 정의는 세타를 -b로 치환한 것. 
- 식의 한편에 세타가 있으면 계속 식을 변경해줘야하니까, 이항해주면서 식을 간편하게 만듦.
- 

### Q2. 편향이 있는 퍼셉트론과 없는 퍼셉트론은 어떤 차이가 있을까요? 
- 편향이 있는 퍼셉트론은 특정 조건에서 강제로 활성화 또는 비활성화 할 수 있지만  
- 편향이 없는 퍼셉트론은 임의로 설정할 수 없다.  


```python
#2.3.1 간단한 구현
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1

AND(0, 0) #0 출력
AND(1, 0)
AND(0, 1)
AND(1, 1)
```

```python
#2.3.2 가중치와 편향 도입
# x1*w1 + x2*w2 = theta 에서 theta를 b(편향)이라고 하고 다시 정리하면
# x1*w1 + x2*w2 - b = 0으로 수식을 바꿀 수 있음. 이를 위 식에 대입

import numpy as np
x = np.array([0, 1]) #입력
w = np.array([0.5, 0.5]) #가중치
b = -0.7 # 편향

w*x #array([0. , 0.5])

np.sum(w*x) #0.5

np.sum(w*x) + b #-0.1999
```

```python
#2.3.3 가중치와 편향 구현하기

# 1) AND 게이트
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
# 2) NAND 게이트
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) #AND게이트에서 가중치(w,b)의 부호를 바꾸면 됨
    b = 0.7
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else:
        return 1

# 3) OR 게이트
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) #AND게이트에서 가중치(w,b)의 부호를 바꾸면 됨 
    b = -0.2                 #
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

## 2.5 다층 퍼셉트론이 출동한다면
---
- XOR 게이트는 NAND, OR, AND 게이트를 사용하여 만들 수 있음. 

```python
#2.5.2 XOR 게이트 구현하기

def XOR(x1, x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

XOR(0, 0) #0를 출력
XOR(1, 0) #1을 출력
XOR(0, 1) #1을 출력
XOR(1, 1) #0을 출력
```
