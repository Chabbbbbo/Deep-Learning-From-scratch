- 클래스란? 객체를 정의하는 설계도다. 객체의 정보를 가진 멤버변수(속성)와 객체의 멤버함수(기능)으로 구성된다. 
- 클래스를 왜 사용하는가? 동일한 객체를 반복생산, 변수와 함수를 은닉하여 정보를 보호하는 캡슐화, 기존 객체를 기반으로 새 객체를 개조하여 생성하는 상속, 변수 또는 메소드가 상황에 따라 다른 결과를 내는 다향성 (오버로드:같은 함수지만 다른 모양 등으로 쓰는 것, 오버라이딩:상속을 받았을때 수정)
